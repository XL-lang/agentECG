```python
import json
from typing import List, Dict, Any
```

```python
path = "test_results.json"

with open(path, "r") as file:
    datas = json.load(file)
```

```python
def check_answer(template_answer:List[str], answer: List[str] , final_res:str):
    final_res = str(final_res).lower()
    template_answer = [str(i).lower().strip() for i in template_answer]
    filtered_answer = [i for i in template_answer if i in final_res]
    answer = [str(i).lower().strip() for i in answer]
    answer_set = set(answer)
    filtered_answer_set = set(filtered_answer)
    if answer_set.issubset(filtered_answer_set)&(len(answer_set)==len(filtered_answer_set)):
        return  True
    else:
        return False
    
        
```

```python
res = {}
for data in datas:
    question_type = data["question_type"]
    if isinstance(question_type, list):
        question_type = question_type[0]
    answer = data["answer"]
    final_res = data["final_res"]
    template_answer = data["template_answer"]
    tf = check_answer(template_answer, answer, final_res)
    if question_type not in res.keys():
        res[question_type]= {'correct':0, 'incorrect':0, "rate":0}
    if tf:
        res[question_type]["correct"] += 1
    else:
        res[question_type]["incorrect"] += 1
    res[question_type]["rate"] = res[question_type]["correct"] / (res[question_type]["correct"] + res[question_type]["incorrect"])
        
```
