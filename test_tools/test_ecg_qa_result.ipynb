```python
import json
```

```python
path = "test_results.json"

with open(path, "r") as file:
    datas = json.load(file)
datas[101]
```

```python
def check_answer(question_type:str, answer: str , final_res:str):
    valid = False
    correct = False
    if "verify" in question_type:
        valid = True
        if answer[0] == "no":
            if "no" in final_res.lower():
                correct = True
        elif answer[0] == "yes":
            if "yes" in final_res.lower():
                correct = True
        else:
            valid = False
    if "choose" in question_type:
      
        if answer[0] in final_res.lower():
            valid = True
            correct = True
            print(f"✅Answer: {answer[0]}, Final_res: {final_res}")
        else:
            valid = True
            print(f"❌Answer: {answer[0]}, Final_res: {final_res}")
    return valid, correct
        
```

```python
res = {}
for data in datas:
    question_type = data["sample"]["question_type"][0]
    answer = data["sample"]["answer"]
    final_res = data["final_res"]
    if isinstance(final_res, bool):
        if final_res:
            final_res = "yes"
        else:
            final_res = "no"
    valid, correct = check_answer(question_type, answer, final_res)
    if question_type not in res.keys():
        res[question_type] = {"valid": 0, "invalid":0 ,"correct": 0}
    if valid:
        res[question_type]["valid"] += 1
    else:
        res[question_type]["invalid"] += 1
    if correct:
        res[question_type]["correct"] += 1
```
