```python
from data.EcgQaData import ECGQAClassificationDataset,ECGQAClassificationManager, EcgQaDataset,Special_query,QueryType
from dataset.config import ecg_qa_types
import pandas as pd
from models.pytorch_inception.ecgqa_classifier import ECGQAClassifierManager
import ast
from typing import List
ids = [8]
count_dict = {}
for id in ids:
    count_dict[str(id)] = [0,0]

def compare_result(result:List[str],answer:List[str])->bool:
    result_set = set(result)
    answer_set = set(answer)
    return result_set == answer_set
special_queries = [Special_query(QueryType.template_id_query, "8")]
eqc = EcgQaDataset(dataset_dir="dataset/ecgqa_ptbxl/template/valid/",sample_limit=5000,special_queries=special_queries)
for sample in eqc:
    id = int(sample["template_id"])
    if str(id) not in count_dict:
        continue
    datas = sample["ecg_datas"]
    answer = sample["answer"]
    eqcm = ECGQAClassifierManager(datas)
    predict_result = eqcm.predict(id)
    key_list: List[str] = [i[0][0] for i in predict_result]  # pyright: ignore[reportAssignmentType]
    if compare_result(key_list,answer):
        count_dict[str(id)][0] += 1
    else:
        count_dict[str(id)][1] += 1








```

```python
value = list(count_dict.values())[0]
print(f"true: {value[0]}, false: {value[1]},rate: {value[0]/(value[0]+value[1])}")
```
